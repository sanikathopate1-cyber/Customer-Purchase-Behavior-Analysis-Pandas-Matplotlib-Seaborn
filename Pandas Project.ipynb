```python
df["sales"] = pd.to_numeric(df['sales'], errors='coerce')
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True)
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='category', y='sales', data=df, palette='viridis')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

2. Profit by Region (Bar Chart)

```python
plt.figure(figsize=(7,4))
sns.barplot(x='region', y='profit', data=df, palette='coolwarm')
plt.title("Profit by Region")
plt.xticks(rotation=30)
plt.show()
```

3. Discount vs Profit (Scatter Plot)

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='discount', y='profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

4. Yearly Sales Trend (Line Chart)

```python
yearly_sales = df.groupby('year')['sales'].sum()
plt.plot(yearly_sales, marker='o', color='teal')
plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

5. Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['sales','profit','discount','quantity']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

6.Top 10 Sub-Categories by Sales

```python
top_sub = df.groupby('sub_category')['sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
top_sub.plot(kind='bar')
plt.title("Top 10 Sub-Categories by Sales")
plt.xticks(rotation=45)
plt.show()
```

7. Delivery Time by Ship Mode

```python
df['delivery_days'] = (df['ship_date'] - df['order_date']).dt.days
plt.figure(figsize=(8,6))
sns.boxplot(x='ship_mode', y='delivery_days', data=df, palette='Set2')
plt.title("Delivery Time by Ship Mode")
plt.show()
```
